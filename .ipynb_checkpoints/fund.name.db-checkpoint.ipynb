```python
import requests
import json
import re
```

```python
httpRequestForFundNameDictionary = requests.get("http://fund.eastmoney.com/data/rankhandler.aspx?op=dy&dt=kf&ft=all&rs=&gs=0&sc=qjzf&st=desc&sd=2020-04-28&ed=2021-04-28&es=0&qdii=&pi=2&pn=50&dx=0&v=0.04169829522786084", headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36', 'Referer': 'http://fund.eastmoney.com/data/diyfundranking.html'})
```

```python
rawJsonStringForFundNameDictionary = httpRequestForFundNameDictionary.content.decode("utf-8")
```

```python
tidyJsonStringForFundNameDictionary = re.sub(r'^.*?{', '{', rawJsonStringForFundNameDictionary)
tidyJsonStringForFundNameDictionary = re.sub(r';$', '', tidyJsonStringForFundNameDictionary)
tidyJsonStringForFundNameDictionary = re.sub(r'(\b\w+\b):', r'"\1":', tidyJsonStringForFundNameDictionary)
print(tidyJsonStringForFundNameDictionary)
```

```python
rawFundNameList = json.loads(tidyJsonStringForFundNameDictionary)['datas']
```

```python
tidyFundNameDictionary = {}
for rawFundNameString in rawFundNameList:
    tidyFundNameObject = re.split(r',', rawFundNameString)
    tidyFundNameDictionary[tidyFundNameObject[0]] = tidyFundNameObject[1]
    print(tidyFundNameObject[0], tidyFundNameObject[1])
```
